```python
from functools import reduce

def maximum(x, y):
    return x if x > y else y

numbers = [62,55,45,96,4]
print(reduce(maximum, numbers))
```

```python
def even_product(lst):
    product = 1
    for i in lst:
        if i % 2 == 0:
            product *= i
    return product

print(even_product([2, 3, 4, 5, 6, 7, 8]))
```

```python
def add_three(lst):
    for i in lst:
        if i % 3 == 0:
            print(i + 3)
        else:
            print(i)

numbers = [3, 5, 6, 8, 9, 10, 12]
add_three(numbers) 
```
