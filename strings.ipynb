```python
#Write a python program using inbuilt function to Remove all consonants from each string
def remove(word):
    s = ""
    for i in word:
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
            s = s + i
    return s

words = ["charan","Meghansh","Aryan","iron man"]
for i in words:
    print(remove(i))
```

```python
#Write a python program using inbuilt function to Filter strings that alternate vowels and consonants.
vowels="aeiouAEIOU"
def alternate(st):
    for i in range(len(st)-1):
        if  (st[i] in vowels and st[i+1] in vowels) or (st[i] not in vowels and st[i+1] not in vowels):
            return False
    return True
lists=["cagi", "abdd", "isedo"]
print(list(filter(alternate,lists)))
```

```python
#Write a python program using inbuilt function to Keep strings that contain at least 3 different vowels.
vowels="aeiouAEIOU"
st = ["education", "apple", "orange", "idea"]
for s in st:
    vow = ""
    for l in s:
        if l in vowels and l not in vow:
            vow += l

    if len(vow) >= 3:
        print(s)
```

```python
#Write a python program using inbuilt function to Build a string of the first character from each word
def starting(word):
    return word[0]
words = ["apple", "banana", "orange", "grape"]
result = list(map(starting, words))
print(result)
```

```python
#Write a python program using inbuilt function to Reduce list to a string of only digits extracted from all words.
st = ["ab12", "c34", "d5"]
result = ""
for s in st:
    for l in s:
        if l.isdigit():
            result += l

print(result)
```

```python
#Write a python program using inbuilt function to Double the word at every index divisible by 4
words = ["apple", "banana", "orange", "grape", "mango"]

for i in range(len(words)):
    if i % 4 == 0:
        words[i] = words[i] * 2
print(words)
```

```python
#From a sentence, return all words that appear more than once
def repeated(sentence):
    words = sentence.split()

    for i in words:
        if words.count(i) > 1:
            print(i)

sentence = "amazon uses ai for amazon delivery purpose and amazon ha sthere own ai"
repeated(sentence)
```

```python
#Write a python program using inbuilt function to Sort by proximity to a target length (e.g., 5).
st = ["cat", "elephant", "apple", "ball"]
target = 5
print(sorted(st, key=lambda x: abs(len(x)-target)))
```

```python
#Write a python program using inbuilt function to Filter strings that match their sorted version (alphabetical).
st = ["abc", "cba", "aab", "xyz"]
for s in st:
    if s == "".join(sorted(s)):
        print(s)
```

```python
#Write a python function to Capitalize all words in a list of strings.
def cap(st):
    result = []
    for i in st:
        result.append(i.capitalize())
    return result
st = ["apple", "banana", "mango"]
print(cap(st))
```

```python
#Write a python function to Convert list of strings to their lengths.
def length(st):
    result = []
    for i in st:
        result.append(len(i))
    return result
st = ["apple", "banana", "mango"]
print(length(st))
```

```python
#Write a python function to Filter out strings that are palindromes.
def palindrome(st):
    result = []
    for i in st:
        if i == i[::-1]:
            result.append(i)
    return result
st = ["madam", "apple", "level", "banana"]
print(palindrome(st))
```

```python
#Write a python function to Find the longest string in a list.
def long(st):
    l = st[0]
    for i in st:
        if len(i) > len(l):
            l = i
    return l
st = ["cat", "elephant", "dog"]
print(long(st))
```

```python
#Write a python function to Create a list of tuples (index, string) for strings starting with 'b'.
def start(li):
    result = []
    for i in range(len(li)):
        if li[i].startswith("b"):
            result.append((i, li[i]))
    return result
li = ["apple", "banana", "ball", "cat"]
print(start(li))
```

```python
#Write a python function to Find index of the longest string.
def index(li):
    ind = 0
    for i in range(len(li)):
        if len(li[i]) > len(li[ind]):
            ind = i
    return ind
li = ["cat", "elephant", "dog"]
print(index(li))
```

```python
#Write a python function to Reverse each string in the list.
def rev(li):
    result = []
    for i in li:
        result.append(i[::-1])
    return result
li = ["apple", "banana", "cat"]
print(rev(li))
```

```python
#Write a python function to Get ASCII values of characters in a string.
def asc(t):
    result = []
    for i in t:
        result.append(ord(i))
    return result
t = "ABC"
print(asc(t))
```

```python
#Write a python function to Filter strings that start with a vowel.
def vo(li):
    result = []
    vowels="aeiouAEIOU"
    for i in li:
        if i[0] in vowels:
            result.append(i)
    return result
li = ["apple", "banana", "orange", "umbrella", "cat"]
print(vo(li))
```
