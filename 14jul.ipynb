```python
def even_product(lst):
    product = 1
    for i in lst:
        if i % 2 == 0:
            product *= i
    return product

print(even_product([2, 3, 4, 5, 6, 7, 8]))
```

```python
# 1. Take a list of numbers, and filter out the odd numbers and then double the odd numbers, after which reduce them to a single by subtracting each number
from functools import reduce
numbers = [6,4,5,5,2,1,2]
odd = list(filter(lambda x: x % 2 != 0, numbers))
double = list(map(lambda x: x * 2, odd))
result = reduce(lambda x, y: x - y, double)
print(result)
```

```python

# 7. Write a python program to Sort a list of numbers based on the last digit.
def pre_square(n):
    for i in range(1, n + 1):
        if i * i == n:
            return True
    return False

def remove_squares(lst):
    result = []

    for i in lst:
        if not pre_square(i):
            result.append(i)

    return result

numbers = [4, 7, 9, 10, 16, 18, 25]
print(remove_squares(numbers))
```

```python
# 8. Write a python program to Find the maximum number in the list.
def maximum(lst):
    max = lst[0]
    for i in lst:
        if i > max:
            max = i
    return max
numbers = [10, 25, 8, 40, 15]
print(maximum(numbers))
```

```python
# 9. Write a python program to Find the GCD of all numbers in a list.
from functools import reduce

def gcd(x, y):
    while y != 0:
        x, y = y, x % y
    return x

lst = [52,85,14,12,2,55]

result = reduce(gcd, lst)

print(result)
```

```python
# 10. Write a python program to Find the index of the first number greater than 50.
def first_index(lst):
    for i in range(len(lst)):
        if lst[i] > 50:
            return i

numbers = [20, 35, 60, 45, 80]

print(first_index(numbers))
```
