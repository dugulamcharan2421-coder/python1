```python
num = int(input("Enter a number: "))

if num == 2 or num == 3 or num == 5 or num == 7:
    print("Prime Number")
elif num % 2 == 0 or num % 3 == 0 or num % 5 == 0 or num % 7 == 0:
    print("Not a Prime Number")
else:
    print("Prime Number")
```
